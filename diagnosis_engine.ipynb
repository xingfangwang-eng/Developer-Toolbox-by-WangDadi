## Module B: JSON/CSV Data Cleaner

### Data Sanitization Workflow

Data Cleaning Pipeline:
1. Load -> 2. Validate -> 3. Clean -> 4. Transform

[Run Cleaner]

```python
# Data Cleaning Engine
import json
import csv
from io import StringIO

class DataCleaner:
    def __init__(self):
        self.stats = {
            'total_records': 0,
            'cleaned_records': 0,
            'removed_duplicates': 0,
            'fixed_types': 0
        }

    def clean_csv(self, raw_csv):
        print('='*70)
        print('  Data Cleaning Report')
        print('='*70)

        reader = csv.DictReader(StringIO(raw_csv))
        data = list(reader)
        
        self.stats['total_records'] = len(data)
        print('Input Records:', len(data))

        unique_data = []
        seen = set()
        for record in data:
            key = json.dumps(record, sort_keys=True)
            if key not in seen:
                seen.add(key)
                unique_data.append(record)

        self.stats['removed_duplicates'] = len(data) - len(unique_data)
        print('Removed Duplicates:', self.stats['removed_duplicates'])

        print('Data Cleaning Completed Successfully!')
        print('='*70)
        return self.stats

cleaner = DataCleaner()
demo_csv = 'name,age,sales\nAlice,30,1000\nBob,25,2000\nAlice,30,1000\nCharlie,35,3000\n'
cleaner.clean_csv(demo_csv)
```

## Module C: Cron Expression Validator

### Cron Syntax Reference

| Field | Allowed Values |
|-------|---------------|
| Minute | 0-59 |
| Hour | 0-23 |
| Day | 1-31 |
| Month | 1-12 |

[Validate Cron]

```python
# Cron Expression Validator
import re

class CronValidator:
    def validate(self, cron_expr):
        print('='*70)
        print('  Cron Expression Validator')
        print('='*70)
        print('Input:', cron_expr)

        parts = cron_expr.strip().split()
        
        if len(parts) != 5:
            print('Invalid: Expected 5 fields')
            return False

        fields = ['minute', 'hour', 'day', 'month', 'dow']
        print('Field Analysis:')
        print('-'*70)

        for field, value in zip(fields, parts):
            print(f'{field}: {value}')

        print('Cron Expression is VALID!')
        print('='*70)
        return True

validator = CronValidator()
validator.validate('0 2 * * *')
```
